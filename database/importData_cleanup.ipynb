```python
#Import Dependencies
import pandas as pd
import sqlite3
from sqlite3 import Error
```

```python
# Read in CSVs from folder structure: <POSITION>_stats/
#List of positions
position = ['QB', 'RB', 'TE', 'WR']
#years 
year = [2018, 2019, 2020, 2021]
#empty list for df names created
df_names =[]
#dempty list to hold dfs
df_list = []

#loop through to read csv files to dfs
for x in position: 
    pos= x.lower()
    for y in year:
        filename = f"advanced-{pos}-metrics{y}.csv"
        filepath = f"../DataBase/Resources/{x}_stats/"
        df = f"{pos}_{y}_df"
        df_names.append(df)
        read_df = pd.read_csv(f"{filepath}{filename}")
        df_list.append(read_df)

      
```

```python
#Check df names
df_names
```

```python
#check data list 
len(df_list)
```

```python
#Add year series to each dataframe
for i in range(0,len(df_list)):
    yr = df_names[i][3:7]
    df_list[i]["Year"] = yr
```

```python
#Check year was added
df_list[1].head()
```

```python
#Create empty DF to concatenate position
qb_df = pd.DataFrame()
rb_df = pd.DataFrame()
te_df = pd.DataFrame()
wr_df = pd.DataFrame()
```

```python
#Use coniditionals to concat dataframs
for i in range(0,len(df_list)):
    #Check for QB name
    if df_names[i][:2] == 'qb': 
        qb_df = pd.concat([df_list[i], qb_df], ignore_index=True)
        print(f"index {i} was added to qb")
    #Check for RB name / Add to DF
    elif df_names[i][:2] == 'rb':
        rb_df = pd.concat([df_list[i], rb_df], ignore_index=True)
        print(f"index {i} was added to rb")
    #Check for TE name / Add to DF
    elif df_names[i][:2] == 'te': 
        te_df = pd.concat([df_list[i], te_df], ignore_index=True)
        print(f"index {i} was added to te")
    #Check for WR name / Add to DF
    elif df_names[i][:2] == 'wr': 
        wr_df = pd.concat([df_list[i], wr_df], ignore_index=True)
        print(f"index {i} was added to wr")
    #Check for mistakes
    else:
        print("Uh Oh!")
```
