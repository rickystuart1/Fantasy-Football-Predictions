```python
#Import Dependencies
import psycopg2
import sqlalchemy
import pandas as pd 
from config import password
```

```python
# Read in CSVs
position = ['QB', 'RB', 'TE', 'WR']
year = [2018, 2019, 2020, 2021]
df_names =[]
df_list = []

for x in position: 
    pos= x.lower()
    for y in year:
        filename = f"fantasy-{pos}-leaders{y}.csv"
        filepath = f"../Resources/{x}_stats/"
        df = f"{pos}_{y}_additional_df"
        df_names.append(df)
        read_df = pd.read_csv(f"{filepath}{filename}")
        df_list.append(read_df)

      
```

```python
#Add year to each dataframe
for i in range(0,len(df_list)):
    yr = df_names[i][3:7]
    df_list[i]["Year"] = yr
```

```python
#Create empty DF to concatenate position
qb_add_df = pd.DataFrame()
rb_add_df = pd.DataFrame()
te_add_df = pd.DataFrame()
wr_add_df = pd.DataFrame()
```

```python
#Use coniditionals to concat dataframs
for i in range(0,len(df_list)):
    #Check for QB name
    if df_names[i][:2] == 'qb': 
        qb_add_df = pd.concat([df_list[i], qb_add_df], ignore_index=True)
        print(f"index {i} was added to qb")
    #Check for RB name / Add to DF
    elif df_names[i][:2] == 'rb':
        rb_add_df = pd.concat([df_list[i], rb_add_df], ignore_index=True)
        print(f"index {i} was added to rb")
    #Check for TE name / Add to DF
    elif df_names[i][:2] == 'te': 
        te_add_df = pd.concat([df_list[i], te_add_df], ignore_index=True)
        print(f"index {i} was added to te")
    #Check for WR name / Add to DF
    elif df_names[i][:2] == 'wr': 
        wr_add_df = pd.concat([df_list[i], wr_add_df], ignore_index=True)
        print(f"index {i} was added to wr")
    #Check for mistakes
    else:
        print("Uh Oh!")
```
